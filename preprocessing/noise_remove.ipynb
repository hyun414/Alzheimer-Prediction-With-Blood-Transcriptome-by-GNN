```python
import csv
import numpy as np
import pandas as pd

f = open("/work/home/nhkim/BiS495/data/ADNI_EI_RIN.csv", "r")
reader = csv.reader(f)
profile = list(reader)
prof = np.array(profile)
```

```python
probe = prof[1:,0]

probe_val = np.unique(probe)
probe_num = probe_val.size
mul_num = probe_num - probe.size

print(probe_num)
```

```python
print(prof[1])
```

```python
gene_id = prof[0:,0]
del_gene = []
print(gene_id)

for i in range(1, 49387):
    if gene_id[i] == 'NA':
        del_gene.append(i)

print(del_gene)
```

```python
print(gene_id[49387])
```

```python
prof_del_ver = np.delete(prof, del_gene, axis = 0)
```

```python
print(prof_del_ver.shape)
print(prof_del_ver[0:,0])
```

```python
ex_del_list = []

expression = np.array(prof_del_ver[1:,1:-1], dtype = np.float64)

print(expression[0])
```

```python
median = 3.897
ex_del_list = []

for i in range(48760):
    cnt = 0
    for ex in expression[i]:
        if ex <= median:
            cnt += 1
    
    if cnt > 100:
        ex_del_list.append(i)

print(ex_del_list)
print(len(ex_del_list))
```

```python
new_prof = np.delete(expression, ex_del_list, axis = 0)

print(new_prof[1])
```

```python
new_gene_id = prof_del_ver[1:,0]
```

```python
print(new_gene_id)
```

```python
new_gi = np.delete(new_gene_id, ex_del_list, axis = 0)
```
